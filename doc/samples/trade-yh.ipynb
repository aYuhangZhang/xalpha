# 指定标的的交易分析

* 想要进行交易分析首先需要交易单

```python
path='../../tests/fund2021.csv'
read = xa.record(path) 
read.status 
```

```python
GTRY_JunGong = xa.fundinfo("001838")
HTF_XinNengYuanCar = xa.fundinfo("501057")
ZhO_YiLiao = xa.fundinfo("003095")
FG_HuGangShen = xa.fundinfo("005847")
```

```python
GTRY_JunGong_t = xa.trade(GTRY_JunGong, read.status) # 给定标的和账单，交易就生成了
HTF_XinNengYuanCar_t = xa.trade(HTF_XinNengYuanCar, read.status) # 给定标的和账单，交易就生成了
ZhO_YiLiao_t = xa.trade(ZhO_YiLiao, read.status) # 给定标的和账单，交易就生成了
FG_HuGangShen_t = xa.trade(FG_HuGangShen, read.status) # 给定标的和账单，交易就生成了
GTRY_JunGong_t.v_totvalue() # 基金持有金额的变动可视化
```
